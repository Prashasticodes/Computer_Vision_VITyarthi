```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Color Spaces Conversion
print("--- 1. Color Spaces Conversion ---")

# Create a dummy BGR image (OpenCV uses BGR by default)
# A simple 100x100 blue square
dummy_bgr = np.zeros((100, 100, 3), dtype=np.uint8)
dummy_bgr[:, :, 0] = 255  # Set Blue channel to max

# Convert BGR to Grayscale
gray_img = cv2.cvtColor(dummy_bgr, cv2.COLOR_BGR2GRAY)

# Convert BGR to HSV
hsv_img = cv2.cvtColor(dummy_bgr, cv2.COLOR_BGR2HSV)

# Convert BGR to RGB (for Matplotlib display, as it expects RGB)
rgb_img = cv2.cvtColor(dummy_bgr, cv2.COLOR_BGR2RGB)

print(f"Original BGR Shape: {dummy_bgr.shape}")
print(f"Grayscale Shape: {gray_img.shape}")
print(f"HSV Shape: {hsv_img.shape}")

# Display results using Matplotlib
plt.figure(figsize=(10, 3))
plt.subplot(1, 3, 1)
plt.imshow(rgb_img)
plt.title('Original (RGB)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(hsv_img)
plt.title('HSV')
plt.axis('off')
plt.show()

# Clean up
plt.close()
```

```python
import cv2
import numpy as np

# 3. Video Processing: Capturing, Writing, Flipped Video
print("\n--- 3. Video Processing ---")

# Define the file to save the video to
output_filename = 'flipped_output.avi'

# Start video capture (0 is usually the default camera)
cap = cv2.VideoCapture(0)

# Check if the camera opened successfully
if not cap.isOpened():
    print("Error: Could not open video stream/file (Webcam not found or busy). Skipping video processing.")
else:
    # Get frame properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    # Define the codec and create VideoWriter object
    # XVID is a common, reliable codec. Adjust based on system compatibility.
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_filename, fourcc, fps, (frame_width, frame_height))

    print(f"Recording and writing video for 5 seconds to {output_filename}...")

    # Process and write video for a set number of frames (e.g., 5 seconds at 30 FPS)
    frames_to_capture = int(5 * fps) if fps > 0 else 150 # Aim for 5 seconds or 150 frames

    frame_count = 0
    while(cap.isOpened() and frame_count < frames_to_capture):
        ret, frame = cap.read()

        if ret:
            # Flip the frame horizontally
            flipped_frame = cv2.flip(frame, 1) # 1 = horizontal flip, 0 = vertical

            # Write the flipped frame to the output file
            out.write(flipped_frame)

            # Show the live feed (optional for lab manual, but helpful for testing)
            cv2.imshow('Flipped Live Feed (Press Q to stop early)', flipped_frame)

            frame_count += 1

            # Break the loop on 'q' key press
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        else:
            break

    # Release everything when the job is finished
    cap.release()
    out.release()
    cv2.destroyAllWindows()
    print(f"Video successfully recorded and saved as {output_filename}.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 4. Edge Detection (Canny)
print("\n--- 4. Edge Detection (Canny) ---")

# Create a grayscale dummy image with a square in the middle
dummy_img = np.zeros((200, 200), dtype=np.uint8)
dummy_img[50:150, 50:150] = 150 # Grey square
dummy_img[75:125, 75:125] = 255 # White square inside

# Apply Gaussian Blur (recommended before Canny)
blurred = cv2.GaussianBlur(dummy_img, (5, 5), 0)

# Canny Edge Detection
# Threshold 1 and 2: Used for hysteresis thresholding
canny_edges = cv2.Canny(blurred, 100, 200)

# Display results
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(dummy_img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(canny_edges, cmap='gray')
plt.title('Canny Edges')
plt.axis('off')
plt.show()

# Clean up
plt.close()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 5. Histogram Equalization
print("\n--- 5. Histogram Equalization ---")

# Create a low-contrast dummy image (dark tones only)
dummy_img = np.random.randint(0, 100, size=(100, 100), dtype=np.uint8)

# Apply Histogram Equalization
equalized_img = cv2.equalizeHist(dummy_img)

# Calculate Histograms
hist_original = cv2.calcHist([dummy_img], [0], None, [256], [0, 256])
hist_equalized = cv2.calcHist([equalized_img], [0], None, [256], [0, 256])

# Display results
plt.figure(figsize=(10, 8))

# Images
plt.subplot(2, 2, 1)
plt.imshow(dummy_img, cmap='gray')
plt.title('Original Low Contrast Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(equalized_img, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

# Histograms
plt.subplot(2, 2, 3)
plt.plot(hist_original, color='black')
plt.title('Original Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.plot(hist_equalized, color='black')
plt.title('Equalized Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Clean up
plt.close()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 6. Masking
print("\n--- 6. Masking ---")

# Create a dummy image (Blue color)
img = np.zeros((200, 200, 3), dtype=np.uint8)
img[:, :, 0] = 200 # Blue channel

# Create a mask (black background)
mask = np.zeros((200, 200), dtype=np.uint8)

# Draw a white filled circle on the mask (this is the region of interest)
center = (100, 100)
radius = 70
cv2.circle(mask, center, radius, 255, -1) # -1 means fill the circle

# Apply the mask: only the circular region of the image will be kept
masked_img = cv2.bitwise_and(img, img, mask=mask)

# Display results
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(mask, cmap='gray')
plt.title('Mask (Circle)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(masked_img, cv2.COLOR_BGR2RGB))
plt.title('Masked Image')
plt.axis('off')

plt.show()

# Clean up
plt.close()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 7. Morphological Operations
print("\n--- 7. Morphological Operations ---")

# Create a dummy image (white square with some 'salt' noise)
img_size = 100
dummy_img = np.zeros((img_size, img_size), dtype=np.uint8)
dummy_img[20:80, 20:80] = 255 # White square
# Add some salt noise (white pixels)
dummy_img[np.random.randint(0, img_size, 50), np.random.randint(0, img_size, 50)] = 255
# Add some pepper noise (black holes in the square)
dummy_img[30:70, 30:70][np.random.randint(0, 40, 20), np.random.randint(0, 40, 20)] = 0

# Define a kernel (structuring element)
kernel = np.ones((5, 5), np.uint8)

# Operations
erosion = cv2.erode(dummy_img, kernel, iterations=1)
dilation = cv2.dilate(dummy_img, kernel, iterations=1)
opening = cv2.morphologyEx(dummy_img, cv2.MORPH_OPEN, kernel) # Erosion followed by Dilation (removes small white noise/salt)
closing = cv2.morphologyEx(dummy_img, cv2.MORPH_CLOSE, kernel) # Dilation followed by Erosion (removes small black noise/pepper)

# Display results
titles = ['Original', 'Erosion', 'Dilation', 'Opening', 'Closing']
images = [dummy_img, erosion, dilation, opening, closing]

plt.figure(figsize=(15, 4))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()

# Clean up
plt.close()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 8. K-means Algorithm
print("\n--- 8. K-means Algorithm ---")

# A. Data Clustering Example (Synthetic Data)
print("A. K-means on Synthetic Data:")
X = np.array([[1, 2], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11]])

kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)
kmeans.fit(X)

print("Cluster labels for data points:", kmeans.labels_)
print("Cluster Centers:\n", kmeans.cluster_centers_)

# B. K-means for Image Segmentation (Using a simple image)
print("\nB. K-means for Image Segmentation:")

# Create a simple BGR image with three distinct color regions
img = np.zeros((50, 50, 3), dtype=np.uint8)
img[:, 0:17] = [255, 0, 0]    # Blue
img[:, 17:34] = [0, 255, 0]   # Green
img[:, 34:50] = [0, 0, 255]   # Red

# Reshape the image into a list of pixel values (R, G, B)
Z = img.reshape((-1, 3)).astype(np.float32)

# Define criteria and apply K-means
K = 3 # We know there are 3 colors
kmeans_img = KMeans(n_clusters=K, random_state=0, n_init=10)
kmeans_img.fit(Z)

labels = kmeans_img.labels_
centers = kmeans_img.cluster_centers_

# Now convert back to an 8-bit image
center = np.uint8(centers)
res = center[labels.flatten()]
segmented_img = res.reshape((img.shape))

# Display results
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(segmented_img, cv2.COLOR_BGR2RGB))
plt.title('K-means Segmented Image (K=3)')
plt.axis('off')
plt.show()

# Clean up
plt.close()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 9. Watershed Algorithm - Image Segmentation
print("\n--- 9. Watershed Algorithm ---")

# Create a synthetic image of two overlapping white circles
img = np.zeros((150, 150), dtype=np.uint8)
cv2.circle(img, (50, 75), 40, 255, -1) # Left circle
cv2.circle(img, (100, 75), 40, 255, -1) # Right circle
img = cv2.GaussianBlur(img, (9, 9), 0) # Smooth for better gradient

# Convert to 3-channel BGR for visualization
img_color = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

# Threshold to get a binary image
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
thresh = np.uint8(thresh)

# Find sure foreground area (Distance Transform)
D = cv2.distanceTransform(thresh, cv2.DIST_L2, 5)
# Simple threshold for the sure foreground (peaks of the distance map)
ret, sure_fg = cv2.threshold(D, 0.5 * D.max(), 255, 0)
sure_fg = np.uint8(sure_fg)

# Find unknown region
sure_bg = cv2.dilate(thresh, np.ones((3, 3), np.uint8), iterations=3)
unknown = cv2.subtract(sure_bg, sure_fg)

# Marker labelling
ret, markers = cv2.connectedComponents(sure_fg)

# Add one to all labels so that sure background is 1 instead of 0
markers = markers + 1

# Mark the unknown region with 0
markers[unknown == 255] = 0

# Apply the Watershed algorithm
markers = cv2.watershed(img_color, markers)
img_color[markers == -1] = [0, 0, 255] # Mark boundary in red

# Create a segmented output image
segmented_output = np.zeros_like(img_color)
for i in range(1, ret + 1):
    # Assign a random color to each segmented region
    segmented_output[markers == i] = np.random.randint(0, 256, 3)

# Display results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Overlapping Shapes')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(segmented_output, cv2.COLOR_BGR2RGB))
plt.title('Watershed Segmented Output')
plt.axis('off')

plt.show()

# Clean up
plt.close()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 10. Object Detection (Simple Template Matching)
print("\n--- 10. Object Detection (Template Matching) ---")

# Create a simple main image (a white background with one large red square)
main_img = np.zeros((200, 200, 3), dtype=np.uint8)
main_img[:, :] = [255, 255, 255] # White
main_img[50:150, 50:150] = [0, 0, 255] # Red square (BGR)

# Create the template (a small red square, the "object" to detect)
template = np.zeros((40, 40, 3), dtype=np.uint8)
template[:, :] = [0, 0, 255] # Red square (BGR)

# Template Matching
# Result is a grayscale image where bright spots indicate a match
result = cv2.matchTemplate(main_img, template, cv2.TM_CCOEFF_NORMED)

# Find the best match location
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

# Get the coordinates and size of the template
top_left = max_loc
h, w, c = template.shape
bottom_right = (top_left[0] + w, top_left[1] + h)

# Draw a bounding box around the detected object on the main image
detection_img = main_img.copy()
# Draw a green rectangle for the detected object
cv2.rectangle(detection_img, top_left, bottom_right, (0, 255, 0), 2) # (0, 255, 0) is Green

# Display results
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(main_img, cv2.COLOR_BGR2RGB))
plt.title('Main Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(template, cv2.COLOR_BGR2RGB))
plt.title('Template (Object)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(detection_img, cv2.COLOR_BGR2RGB))
plt.title(f'Detected Object at {top_left}')
plt.axis('off')
plt.show()

# Clean up
plt.close()
```
